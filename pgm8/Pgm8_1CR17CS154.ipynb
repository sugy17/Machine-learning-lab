```python
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.cluster import KMeans
import sklearn.metrics as sm
from sklearn.metrics import accuracy_score, confusion_matrix
import pandas as pd
import numpy as np
iris = datasets.load_iris()
print("Iris Data: \n", iris.data)
print("\n\nIris Features: \n", iris.feature_names)
print("\n\nIris Target: \n", iris.target)
print("\n\nIris Target Names: \n", iris.target_names)
X = pd.DataFrame(iris.data)
print(X)
```

```python
X.columns = ['Sepal_Length', 'Sepal_Width', 'Petal_Length', 'Petal_Width']
y = pd.DataFrame(iris.target)
y.columns = ['Targets']
plt.figure(figsize = (14, 7))
colormap = np.array(['red', 'lime', 'black'])
plt.subplot(1, 2, 1)
plt.scatter(X.Sepal_Length, X.Sepal_Width, c=colormap[y.Targets], s=40)
plt.title('Sepal')
plt.subplot(1, 2, 2)
plt.scatter(X.Petal_Length, X.Petal_Width, c=colormap[y.Targets], s=40)
plt.title('Petal')
```

```python
model = KMeans(n_clusters=3)
model.fit(X)
```

```python
model.labels_
```

```python
sm.accuracy_score(y, model.labels_)
```

```python
sm.confusion_matrix(y, model.labels_)
```

```python
plt.figure(figsize=(14, 7))
colormap = np.array(['orange', 'blue', 'grey'])
plt.subplot(1, 2, 1)
plt.scatter(X.Petal_Length, X.Petal_Width, c=colormap[y.Targets], s=40)
plt.title("Real Classification")
plt.subplot(1, 2, 2)
plt.scatter(X.Petal_Length, X.Petal_Width, c=colormap[model.labels_], s=40)
plt.title("K Mean Classification")
```

```python
from sklearn import preprocessing
scaler = preprocessing.StandardScaler()
scaler.fit(X)
xsa = scaler.transform(X)
xs = pd.DataFrame(xsa, columns=X.columns)
xs.sample(5)
```

```python
from sklearn.mixture import GaussianMixture
gmm = GaussianMixture(n_components=3)
gmm.fit(xs)
```

```python
y_cluster_gmm = gmm.predict(xs)
y_cluster_gmm
```

```python
plt.subplot(1, 2, 1)
plt.scatter(X.Petal_Length, X.Petal_Width, c=colormap[y_cluster_gmm], s=40)
plt.title('GMM Classification')
```

```python
sm.accuracy_score(y, y_cluster_gmm)
```

```python
sm.confusion_matrix(y, y_cluster_gmm)
```
